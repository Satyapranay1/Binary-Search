```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
data.head()
```

```python
data.team1.unique()
```

```python
data.venue.unique()
```

```python
data = data.drop(['id','season','date','match_type','player_of_match','result','result_margin','target_runs','target_overs','super_over','method','umpire1','umpire2','Unnamed: 20','city'],axis = 1)
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data.fillna('ffill',inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['venue'] = le.fit_transform(data['venue'])
data['team1'] = le.fit_transform(data['team1'])
data['team2'] = le.fit_transform(data['team2'])
data['toss_winner'] = le.fit_transform(data['toss_winner'])
data['toss_decision'] = le.fit_transform(data['toss_decision'])
data['winner'] = le.fit_transform(data['winner'])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
ct = ColumnTransformer([('encoder',OneHotEncoder(),[0])],remainder = 'passthrough')
ct.fit_transform(data)
enc = OneHotEncoder()
```

```python
data.head()
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['winner'],axis = 1)
y = data['winner']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
```

```python
pred1 = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,pred1) * 100)
```

```python
report = classification_report(y_test,pred1)
print(report)
```
